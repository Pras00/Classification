Import Library

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import itertools
```

### **1. Ulangi langkah pada kode 1 untuk dataset iris dan adult. Masing masing gunakan perbandingan 0.2, 0.3, 0.4. Tuliskan akurasinya**

**Import Dataset Iris**

```python
data_iris = pd.read_csv("iris.csv")
data_iris
```

**Membuat Dataframe baru**

```python
inputs = data_iris.drop(["Species", "Id"],axis='columns')
target = data_iris["Species"]
inputs
```

Membuat DataFrame baru yang disimpan dalam variabel inputs. DataFrame ini berisi semua atribut dari data_iris kecuali "Species" dan "Id". Argumen axis='columns' menunjukkan bahwa yang dihapus adalah kolom, bukan baris. <br>
Lalu kita menyimpan isi dari tiap kolom "Species ke dalam variabel target. Dalam konteks ini, "Species" berperan sebagai target atau label yang akan diprediksi dalam tugas klasifikasi.

**Membuat Dataframe dummies baru**

```python
dummies = pd.get_dummies(data_iris["PetalLengthCm"])
dummies
```

Membuat DataFrame baru yang disimpan dalam variabel dummies. DataFrame ini menggunakan metode get_dummies dari pandas untuk melakukan one-hot encoding pada kolom "PetalLength" dari data_iris. One-hot encoding digunakan untuk mengubah variabel kategorikal menjadi representasi biner (1 dan 0) dengan kolom-kolom baru yang mewakili setiap nilai kategorikal. Setiap kolom baru akan memiliki nilai 1 jika data pada baris tersebut sesuai dengan nilai kategorikal yang diwakilinya, dan 0 jika tidak sesuai.

**Menggabungkan Dataframe Dummies ke dalam dataframe Iris**

```python
inputs = pd.concat([data_iris, dummies],axis='columns')
inputs
```

Variabel inputs akan berisi DataFrame yang merupakan hasil penggabungan antara data_iris dan dummies, sehingga memiliki semua atribut dari data_iris serta kolom-kolom baru yang mewakili one-hot encoding dari kolom "PetalLength" (dataframe Dummies).

**Menghapus kolom Species**

```python
inputs.drop(['Species', 'PetalLengthCm'],axis='columns',inplace=True)
inputs
```

kolom 'Species', 'PetalLength', '1.5' akan dihapus dari DataFrame inputs, sehingga DataFrame inputs akan menjadi DataFrame yang hanya berisi atribut-atribut yang digunakan sebagai input untuk pelatihan model.

**Mencari Nilai Kolom yang bersifat Null atau kosong**

```python
inputs.columns[inputs.isna().any()]
```

Kita melakukan cek di beberapa kolom dataframe inputs apakah terdapat nilai null atau NaN. Jika ada maka akan ditampilkan sebagai output nama kolom yang meiliki nilai null atau NaN. <br>
Karena hasil output tidak ditemukan nilai null pada masing - masing kolom, maka tidak perlu mencari rata - rata dari nilai kolom untuk mengisi kolom null atau NaN-nya

##### **Test Size = 0.2** <br>
**Menguji dataset dengan data training sebesar 80% dan data testing sebesar 20% dari dataset awal.**

```python
# Membagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

- inputs: Variabel yang berisi atribut-atribut (features) yang digunakan sebagai masukan (X) dalam pemodelan.
- target: Variabel yang berisi target (label) yang ingin diprediksi (y).
- test_size=0.2: Parameter yang menentukan ukuran (proporsi) data uji dalam pemisahan. Nilai 0.2 menandakan bahwa data uji akan memiliki ukuran 20% dari keseluruhan dataset, sedangkan data latih akan memiliki ukuran 80%.
- X_train, X_test, y_train, y_test: Variabel-variabel yang menyimpan hasil pemisahan data menjadi data latih dan data uji. X_train dan y_train akan berisi data latih untuk atribut dan target, sedangkan X_test dan y_test akan berisi data uji untuk atribut dan target.

**Menginisialisasi atau Membuat model Gaussian Naive Bayes**

```python
# Inisialisasi model Naive Bayes
model = GaussianNB()
```

Dengan menggunakan GaussianNB(), kita membuat objek model Gaussian Naive Bayes yang akan digunakan untuk melakukan pelatihan dan prediksi pada dataset. Objek model ini akan memiliki metode dan atribut yang dapat digunakan untuk melatih model dengan data latih dan melakukan prediksi terhadap data uji.

**Melatih model Gaussian Naive Bayes dengan menggunakan data latih X_train dan label y_train**

```python
# Melatih model menggunakan data latih
model.fit(X_train.values, y_train)
```

melatih model Gaussian Naive Bayes dengan menggunakan data latih X_train dan label y_train. Proses pelatihan akan menghasilkan model yang siap digunakan untuk melakukan prediksi pada data baru.

```python
X_test[0:10]
```

Menampilkan subset dari data uji X_test yang terdiri dari 10 baris pertama

```python
y_test[0:10]
```

Menampilkan subset dari data uji y_test yang terdiri dari 10 baris pertama

**Melakukan Prediksi**

```python
# Melakukan prediksi pada data uji
model.predict(X_test.values[0:10])
```

Melakukan prediksi menggunakan model yang telah dilatih (model) terhadap subset data uji (X_test) yang terdiri dari 10 data pertama (X_test[0:10]).

**Melakukan Prediksi Probabilitas**

```python
model.predict_proba(X_test.values[:10])
```

Melakukan prediksi probabilitas kelas menggunakan model yang telah dilatih (model) terhadap subset data uji (X_test) yang terdiri dari 10 data pertama (X_test[:10]).

**Melakukan Perhitungan Score**

```python
model.score(X_test.values,y_test)
```

Metode score() digunakan untuk menghitung akurasi model pada data uji. Pada syntax ini, X_test merupakan dataset atribut yang digunakan sebagai data uji, sedangkan y_test merupakan label atau target yang sesuai dengan data uji. Metode score() akan membandingkan prediksi yang dihasilkan oleh model dengan label sebenarnya pada data uji, dan menghitung persentase kecocokan atau akurasi model.

```python
# Menghitung akurasi prediksi
prediction = model.predict(X_test.values)
accuracy = accuracy_score(y_test, prediction)
print("Akurasi:", accuracy)
```

Dilakukan prediksi menggunakan model yang telah dilatih pada data uji (X_test) menggunakan metode predict(). Kemudian, dilakukan perhitungan akurasi menggunakan fungsi accuracy_score() dengan membandingkan hasil prediksi dengan label sebenarnya (y_test).

Didapat Output yang menunjukkan bahwa model yang digunakan dalam evaluasi memberikan prediksi yang sangat baik untuk setiap contoh dalam data latih. Skor ini menunjukkan bahwa prediksi model memiliki kecocokan yang sangat tinggi dengan label yang sebenarnya pada data latih.

### **2. Untuk masing masing dataset carilah subset kombinasi dari attribut yang memeberi hasil klasifikasi lebih baik dari data asli**

```python
# buat kombinasi subset kolom SepalLengthCm
# load dataset iris
data_iris = pd.read_csv("Iris.csv")

# pisahkan antara input dan target
inputs = data_iris.drop(['Species', 'Id'], axis='columns')
target = data_iris['Species']

subsets = []
for i in range(1, len(inputs.columns)):
    subsets.extend(itertools.combinations(['PetalLengthCm'] + list(inputs.columns[i:]), i))

# lakukan splitting dengan test size 0.2
for subset in subsets:
    X_train, X_test, y_train, y_test = train_test_split(inputs[list(subset)], target, test_size=0.2, random_state=0)
    
    # buat model Gaussian Naive Bayes
    model = GaussianNB()
    
    # lakukan training model
    model.fit(X_train, y_train)
    
    # lakukan prediksi
    prediction = model.predict(X_test)
    
    # hitung akurasi
    accuracy = accuracy_score(y_test, prediction)
    
    # cetak hasil
    print(f"Subset kolom: {subset}")
    print("Akurasi:", accuracy)
    print("")

```
